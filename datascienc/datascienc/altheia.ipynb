```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = r'D:\datascienc\combined_data.csv'
df = pd.read_csv(file_path)
```

```python
# Display the number of rows before removing duplicates
print("Number of rows before removing duplicates:", len(df))

# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()

# Display the number of rows after removing duplicates
print("Number of rows after removing duplicates:", len(df_no_duplicates))
```

```python
# Visualize the distribution of target variables (Diabetes, HyperTension)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='Diabetes', data=df)
plt.title('Distribution of Diabetes')

plt.subplot(1, 2, 2)
sns.countplot(x='HyperTension', data=df)
plt.title('Distribution of HyperTension')

plt.show()
```

```python
# Visualize categorical variables using bar charts
categorical_columns = ['Race']  # Add more categorical columns as needed

for column in categorical_columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=column, data=df)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Create histograms for numerical variables
numerical_columns = ['Age', 'BMI']  # Add more numerical columns as needed

for column in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], bins=20)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Use stacked bar charts for categorical variables against target variables
for column in categorical_columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=column, hue='Diabetes', data=df)
    plt.title(f'Distribution of {column} by Diabetes')
    plt.show()
```

```python
# Numerical variables of interest
numerical_columns_of_interest = ['Age', 'BMI', 'Alcohol_Consumption', 'Smoking_level', 'Physical_activity']

# Target variables
target_variables = ['Diabetes', 'HyperTension']

# Extract the relevant columns from the dataframe
selected_columns = numerical_columns_of_interest + target_variables
selected_df = df[selected_columns]

# Compute correlation coefficients
correlation_matrix = selected_df.corr()

# Visualize correlations using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap between Age, BMI, Alcohol, Smoking, Physical Activity, Diabetes, and Hypertension')
plt.show()
```

```python
# Categorical variables of interest
categorical_columns = ['Alcohol_Consumption', 'Smoking_level', 'Physical_activity']

# Target variables
target_variables = ['Diabetes', 'HyperTension']

# Create stacked bar charts
for target_variable in target_variables:
    for column in categorical_columns:
        plt.figure(figsize=(10, 5))
        sns.countplot(x=column, hue=target_variable, data=df)
        plt.title(f'Distribution of {column} by {target_variable}')
        plt.show()
```
